```python
%matplotlib inline
import matplotlib.pylab as plt #data visualization package
import pandas as pd #data analysis and manipulation tool or package
from osgeo import gdal #GDAL
import numpy as np #numpy for scientific computing
import matplotlib.pyplot as plt
import os, sys 
from skimage import io 
from skimage import exposure

import geopandas as gp
from shapely.geometry import Point
import geoviews as gv
from geoviews import opts, tile_sources as gvts
import holoviews as hv
gv.extension('bokeh', 'matplotlib')


import sys
if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")

os.chdir('//home/jovyan/RSlab3/Lullymore_Bands/Stack') 
```

```python
s2 = gdal.Open('Lully9Band.tif') #Image file
s2 = s2.ReadAsArray()
#s2 = s2.ReadAsArray(xoff = 1000, yoff = 1000, xsize = 1000, ysize = 1000)


def showImage(b1, b2, b3):
    img = np.dstack([b1/b1.max(), b2/b2.max(), b3/b3.max()])
    for i in range(3):
        v_low, v_high = np.percentile(img[:,:,i], (0,98))
        # Rescale the intensity of the input images 0-100 range
        img[:,:,i] = exposure.rescale_intensity(img[:,:,i],in_range=(v_low, v_high)) 
    return img

%matplotlib inline
imageRGB = showImage(s2[0], s2[1], s2[1])
fig, ax = plt.subplots(figsize=(20,10))
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.imshow(imageRGB)
ax.set_title('Study Area')

plt.show()
```

```python
#Bringing In Training Data 
train_data = gp.read_file('TrainingFinal_2157.geojson')


LC_class_names = ['Water', 'Artificial', 'Bare Peat', 'Coniferous Woodland', 'High Vegetation', 'Low Vegetation', 'Pasture'];

# Define a function for visualizing training data points
def pointVisual(features, vdims):
    return (gvts.EsriImagery * gv.Points(features, vdims=vdims).options(tools=['hover'], 
                                                     cmap='tab10', height=800, 
                                                     width=1000, size=5, 
                                                     color='ClassName', 
                                                     fontsize={'xticks': 10, 'yticks': 10, 
                                                        'xlabel':16, 'ylabel': 16}))

# Create a list of geodataframe columns to be included as attributes in the output map
vdims = []
for f in train_data:
    if f not in ['geometry']:
        vdims.append(f)
vdims

#Ignore shapely deprecations warnings
import warnings
warnings.filterwarnings('ignore')

# Call the function for plotting the GEDI points
pointVisual(train_data, vdims = vdims)

print(train_data.crs)
train_data.head()
```

```python
#Training Random Forest
# Get the unique class labels from the training data
class_labels = np.unique(train_data['Class'])

import rasterio
from rasterio.features import geometry_mask
from shapely.geometry import mapping
from rasterio.features import rasterize

# Load the image
with rasterio.open('LillyBandsOBIADSM1.tif') as src:
    s2_data = src.read()
    s2_meta = src.meta
    crs = src.crs 
    print(crs)
    
# Rasterize the training points to a binary mask
train_mask = rasterize(
    [(mapping(point), class_label) for point, class_label in zip(train_data.geometry, train_data['Class'])],
    out_shape=s2_data.shape[-2:], transform=src.transform, fill=-1, dtype='int16')

# Flatten the Sentinel-2 image into a 2D array
s2_data_2d = s2_data.reshape(s2_data.shape[0], -1).T

# Extract the training samples from the flattened image and mask
train_samples = s2_data_2d[train_mask.flatten() != -1]
train_labels = train_mask[train_mask != -1]

from sklearn.model_selection import train_test_split

 # Split into train and validation set
X_train, X_test, y_train, y_test = train_test_split(train_samples, train_labels, test_size = 0.1, random_state=42, stratify=train_labels)

```

```python
from sklearn.ensemble import RandomForestClassifier

# Fit the random forest classifier to the training data
clf = RandomForestClassifier(n_estimators=100, random_state=0)
clf.fit(X_train, y_train)

from sklearn.metrics import accuracy_score

# Predict the validation data labels using the trained classifier
val_pred = clf.predict(X_test)

# Calculate the classification accuracy
accuracy = accuracy_score(y_test, val_pred)

print("Overall accuracy for Random Forest: " + str(100*accuracy) + "%")
```

```python
# Random Forest Classification

# Perform the classification
s2_data_2d_class = clf.predict(s2_data_2d)

# Reshape the classification result to the original image dimensions
s2_data_forest = s2_data_2d_class.reshape(s2_data.shape[1], s2_data.shape[2])

from rasterio.plot import show
from matplotlib.patches import Patch
from matplotlib.colors import ListedColormap

# visualisation of results
# original image
#plt.figure(figsize=(20,16))
#plt.imshow(imageRGB)

#Define the colorbar 
#cmap = plt.cm.plasma
cmap = plt.cm.viridis
# classified image

# Display the color image
plt.figure(figsize=(20,16))
show(s2_data_forest, cmap=cmap)

# Plot your legend
# legend
fig = plt.figure(figsize=(10,1))
ax = fig.add_subplot(111)
# Define the image
plt.imshow(np.hstack((np.zeros((30,100)), np.ones((30,100)), 2*np.ones((30,100)),
                    3*np.ones((30,100)),np.ones((30, 100)) * 4,np.ones((30, 100)) * 5,np.ones((30, 100)) * 6)), cmap = cmap)

ax.text(20, 20, 'Water', fontsize=15,  color='white')
ax.text(115, 20, 'Artificial', fontsize=15,  color='white')
ax.text(205, 20, 'Bare Peat', fontsize=15,  color='black')
ax.text(305, 20, 'Woodland', fontsize=15,  color='black')
ax.text(400, 20, 'High Veg', fontsize=15,  color='black')
ax.text(500, 20, 'Low Veg', fontsize=15,  color='black')
ax.text(600, 20, 'Pasture', fontsize=15,  color='black')

%matplotlib inline
plt.show()

```

```python
#Random Forest Accuracy 

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Compute the confusion matrix
cm1 = confusion_matrix(y_test, val_pred)
#Define the colorbar 
cmap = plt.cm.a
# Plot the confusion matrix as a heatmap
fig, ax1 = plt.subplots(figsize=(16, 16))
heatmap = sns.heatmap(cm1, fmt='g', cmap='viridis', annot=True, ax=ax1)
ax1.set_xlabel('Predicted class', fontweight = 'bold')
ax1.set_ylabel('True class', fontweight = 'bold')
ax1.set_title('Random Forest Confusion Matrix', fontweight = 'bold')
ax1.xaxis.set_ticklabels(LC_class_names)
ax1.yaxis.set_ticklabels(LC_class_names)
 # Show the heatmap to file
heatmapfig = heatmap.get_figure()

from sklearn.metrics import classification_report

# check accuracy of individual classified class and overall accuracy
print(classification_report(y_test, val_pred, target_names=LC_class_names))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

# Define your classifier
clf = RandomForestClassifier(n_estimators=100, random_state=0)

# Perform cross-validation
cv_scores = cross_val_score(clf, X_train, y_train, cv=10)  # Adjust cv value as needed

# Print cross-validation scores
print("Cross-Validation Scores:", cv_scores)

# Calculate mean and standard deviation of cross-validation scores
mean_cv_score = np.mean(cv_scores)
std_cv_score = np.std(cv_scores)
print("Mean CV Score:", mean_cv_score)
print("Standard Deviation CV Score:", std_cv_score)
```

```python
#Random Forest Variable Importance 
# Get the feature importances from the trained classifier
importances = clf.feature_importances_

# Sort the feature importances in descending order
indices = np.argsort(importances)[::-1]

# Get the names of the features
feature_names = ['4 Band Image', 'NDVI', 'NDWI', 'NDRE', 'Elevation', 'OBIA']
#feature_names = ['NDVI', 'NDWI', 'NDRE', 'Elevation']
%matplotlib inline
# Plot the feature importances as a bar plot
plt.bar(range(len(feature_names)), importances[indices])
plt.xticks(range(len(feature_names)), [feature_names[i] for i in indices])
plt.xlabel('Feature', fontweight = 'bold')
plt.ylabel('Importance', fontweight = 'bold')
plt.title('Random Forest Variable Importance')
plt.show()
```
